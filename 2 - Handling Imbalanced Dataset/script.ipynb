```python
import numpy as np
import pandas as pd

np.random.seed(123)

n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples*class_0_ratio)
n_class_1     = n_samples-n_class_0
```

```python
n_class_0, n_class_1
```

```python
# dataframe

class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0]*n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1]*n_class_1
})
```

```python
df =pd.concat([class_0, class_1]).reset_index(drop=True)
df
```

```python
df['target'].value_counts()
```

```python
df_minority = df[df['target']==1]
df_majoirity = df[df['target']==0]
```

```python
from sklearn.utils import resample
df_minority_upsampled = resample(df_minority, 
         replace=True,
         n_samples=len(df_majoirity), 
         random_state=42)
df_minority_upsampled.shape[0]
```

```python
df_upsampled = pd.concat([df_majoirity, df_minority_upsampled])
df_upsampled
```

```python
df_minority = df[df['target']==1]
df_majoirity = df[df['target']==0]

from sklearn.utils import resample
df_majority_downsampled = resample(df_majoirity, 
         replace=False,
         n_samples=len(df_minority), 
         random_state=42)
df_majority_downsampled.shape[0]

df_downsampled = pd.concat([df_majority_downsampled, df_minority])
df_upsampled
```
